```python
#Importing required libraries

import pandas as pd
from langdetect import detect
from sklearn.preprocessing import LabelEncoder
```

```python
# Reading raw CSV file

rawData = pd.read_csv("F:\Circus Social ML Task\Data\Twitter.csv")
```

# Data Analysis & Preprocessing

```python
#Preprocessing - The process has been documented in the document - "Execution (Raw Plan)"

rawData.drop(['Unnamed: 0','Id'], axis = 1, inplace = True)

rawData.set_index('Published DateTime', inplace = True)
rawData.index = pd.to_datetime(rawData.index)

rawData.rename(columns={"Post Contet": "Post Content"}, inplace = True)
```

## Converting object data types to numeric

```python
#This will take a lot of time to run because of the detect function. Please use the processed data.

langList = []
numWords = []
numNewLines = []

for i in range(0,len(rawData)):
    
    #In case the language is not identifies -> Error handling.
    try: 
        langList.append(detect(rawData['Post Content'][i]))
    except:
        langList.append('Unknown')
        
    numWords.append(len(rawData['Post Content'][i].split()))
    numNewLines.append(len(rawData['Post Content'][i].splitlines()))
```

```python
#Deleting "Post Content" columnn and adding extracted features

rawData.drop(['Post Content'], axis = 1, inplace = True)

rawData.insert(0, 'Language', langList)
rawData.insert(0, 'Word Count', numWords)
rawData.insert(0, 'Line Count', numNewLines)
```

```python
#2 columns have object data type, converting them to numeric

le = LabelEncoder()
rawData['Language']= le.fit_transform(rawData['Language'].values)

rawData.insert(16, 'Media Type 1', pd.get_dummies(rawData['Media Type'], drop_first= True).iloc[:,0].tolist())
rawData.insert(16, 'Media Type 2', pd.get_dummies(rawData['Media Type'], drop_first= True).iloc[:,1].tolist())

# Adding one more feature based on time of the day 
rawData.insert(16, 'Hour', rawData.index.hour)

rawData.drop('Media Type', axis = 1 , inplace = True)
```

# Final Processed Data

```python
rawData.sort_index(inplace = True)
```

```python
rawData.to_csv("F:\Circus Social ML Task\Data\processedData.csv")
```
